```python
raw_data_no_space = pd.read_csv('bank_full_no_space.csv', sep=',')
raw_data_no_space.shape
```

```python
# stip space from values in housing
raw_data_no_space['housing'] = raw_data_no_space['housing'].str.strip()
raw_data_no_space['y'] = raw_data_no_space['y'].str.strip()
```

```python
data = raw_data_no_space.copy()
```

## Data EDA

```python
_ = msno.matrix(data)
```

```python
#list(data)
```

```python
def map_target(row):
    if row['y'] in 'yes':
        return 1
    if row['y'] in 'no':
        return 0
```

```python
data['target'] = data.apply(lambda row: map_target(row), axis=1)
```

```python
print(data['target'].value_counts())
print(data['y'].value_counts())
```

```python
data = data.drop('y',axis=1)
```

```python
# ommit 'day' because it is not significant 

categorical = ['job','marital','education','contact','poutcome','month']

binary = ['default','housing','loan']

continous = ['age','balance','campaign','pdays','previous','duration']

dummy_features = categorical + binary
target = ['target']
```

```python
count_features = categorical + binary + continous + target
len(count_features)
```

```python
for cat in categorical:
    values = data[cat].value_counts()
    print(values)
```

```python
for cat in categorical:
    plt.figure(figsize=(12, 7))
    sns.countplot(x=cat, hue='target' , data=data,palette='RdBu_r')
    plt.show()
```

```python
for bins in binary:
    values = data[bins].value_counts()
    print(values)
```

```python
for bins in binary:
    plt.figure(figsize=(12, 7))
    sns.countplot(x=bins, hue='target' , data=data,palette='RdBu_r')
    plt.show()
```

```python
data_dummies = pd.get_dummies(data[dummy_features],drop_first=True)
```

```python
data.drop(dummy_features,inplace=True,axis=1)
```

```python
data_2 = pd.concat([data,data_dummies],axis=1)
```

```python
#list(data_2)
```

```python
X = data_2.drop(target,axis=1)
y = data_2[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

# Logistic Regression

    I could have used several different models to predict my target feature. Here are a couple of reasons why I used logistic regression over the other models. The first reason is that the target is dichotomous. That's just a fancy way of saying that the target is binary meaning that there are only two possible outcomes, "yes" or "no". Another reason is that the simplicity of a logistic regression model lends itself to industries where there may be regulatory constraints. These constraints may require that the model's function and performance would be explainable in layman's terms. Since this is a bank I took that into account.

<img src="images/linear_vs_logistic.jpg">

```python
logreg = LogisticRegression(solver='liblinear')
logreg.fit(X_train,y_train.values.ravel())
```

```python
predictions = logreg.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
cm = confusion_matrix(y_test,predictions)
TP = cm[1][1]
TN = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]
```

```python
tpr = TP/(TP + FN)
fpr = FN /(FP + TN)
```

```python
y_true = ["0", "1"]
y_pred = ["0", "1"]
data = confusion_matrix(y_test,predictions)
df_cm = pd.DataFrame(data, columns=np.unique(y_true), index = np.unique(y_true))
df_cm.index.name = 'Actual'
df_cm.columns.name = 'Predicted'
plt.figure(figsize = (10,7))
sns.set(font_scale=1.4)#for label size
_ = sns.heatmap(df_cm, cmap="Oranges", annot=True,fmt='g',annot_kws={"size": 16})# font size
```

```python
probs = logreg.predict_proba(X_test)
```

```python
probs = probs[:, 1]
```

```python
auc = roc_auc_score(y_test, probs)
print('AUC: %.2f' % auc)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, probs)
```

```python
def plot_roc_curve(fpr, tpr):
    plt.plot(fpr, tpr, color='orange', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()
```

```python
plot_roc_curve(fpr, tpr)
```
